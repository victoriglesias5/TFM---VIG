```python
from sentence_transformers import SentenceTransformer

# https://huggingface.co/thenlper/gte-large
embedding_model = SentenceTransformer("thenlper/gte-large")


def get_embedding(text: str) -> list[float]:
    if not text.strip():
        print("Attempted to get embedding for empty text.")
        return []

    embedding = embedding_model.encode(text)

    return embedding.tolist()
```

```python
for documento in collection.find():
    posicion = documento.get('Resumen')  # Reemplaza 'posición' con el nombre del campo adecuado
    if posicion:
        embedding = get_embedding(posicion)
        collection.update_one({'_id': documento['_id']}, {'$set': {'Resumen_embedding': embedding}})
```

```python
for documento in collection.find():
    posicion = documento.get('Nombre')  # Reemplaza 'posición' con el nombre del campo adecuado
    if posicion:
        embedding = get_embedding(posicion)
        collection.update_one({'_id': documento['_id']}, {'$set': {'Nombre_embedding': embedding}})
```

```python
def vector_search(user_query, collection):
    """
    Perform a vector search in the MongoDB collection based on the user query.

    Args:
    user_query (str): The user's query string.
    collection (MongoCollection): The MongoDB collection to search.

    Returns:
    list: A list of matching documents.

    """

    # Generate embedding for the user query
    query_embedding = get_embedding(user_query)

    if query_embedding is None:
        return "Invalid query or embedding generation failed."

    # Define the vector search pipeline
    pipeline1= [
        {
            "$vectorSearch": {
                "index": "Resumen_index",
                "queryVector": query_embedding,
                "path": "Resumen_embedding",
                "numCandidates": 150,  # Number of candidate matches to consider
                "limit": 4,  # Return top 4 matches
            }
        },
        {
            "$project": {
                "_id": 0,  # Exclude the _id field
                "Nombre": 1,  # Include the plot field
                "Edad": 1,
                "Pierna Buena": 1,
                "Posición": 1,
                "Equipo": 1,
                "Goles": 1,
                "Asistencias": 1,
                "Proyección": 1,
                "Porterías a Cero": 1,
                "Situación": 1,
                "Comentarios del Ojeador": 1,
                "score": {"$meta": "vectorSearchScore"},  # Include the search score
            }
        },
    ]

    pipeline2= [
        {
            "$vectorSearch": {
                "index": "Nombre_index",
                "queryVector": query_embedding,
                "path": "Nombre_embedding",
                "numCandidates": 150,  # Number of candidate matches to consider
                "limit": 4,  # Return top 4 matches
            }
        },
        {
            "$project": {
                "_id": 0,  # Exclude the _id field
                "Nombre": 1,  # Include the plot field
                "Edad": 1,
                "Pierna Buena": 1,
                "Posición": 1,
                "Equipo": 1,
                "Goles": 1,
                "Asistencias": 1,
                "Proyección": 1,
                "Porterías a Cero": 1,
                "Situación": 1,
                "Comentarios del Ojeador": 1,
                "score": {"$meta": "vectorSearchScore"},  # Include the search score
            }
        },
    ]

    # Execute the search for index 1
    results = list(collection.aggregate(pipeline1))
    #results2 = list(collection.aggregate(pipeline2))

    # Combine and return the results with the best score
    all_results = results #+ results2
    all_results.sort(key=lambda x: x['score'], reverse=True)

    return all_results

    # # Execute the search
    # results = collection.aggregate(pipeline)
    # return list(results)
```

```python
def get_search_result(query, collection):

    get_knowledge = vector_search(query, collection)

    if not get_knowledge:
            return "No solutions found."

    # Return only the top solution
    #top_result = get_knowledge[0]
    search_result = ""
    for top_result in get_knowledge:
      search_result += f"Nombre: {top_result.get('Nombre', 'N/A')}, Edad: {top_result.get('Edad', 'N/A')}, Equipo: {top_result.get('Equipo', 'N/A')}, Pierna: {top_result.get('Pierna Buena', 'N/A')}, Posición: {top_result.get('Posición', 'N/A')}, Goles: {top_result.get('Goles', 'N/A')}, Asistencias: {top_result.get('Asistencias', 'N/A')}, Porterías a cero: {top_result.get('Porterías a Cero', 'N/A')}, Situación: {top_result.get('Situación', 'N/A')}, Proyección: {top_result.get('Proyección', 'N/A')}, Comentarios del Ojeador: {top_result.get('Comentarios del Ojeador', 'N/A')}\n"

    # search_result = ""
    # for result in get_knowledge:
    #     search_result += f"Nombre: {top_result.get('Nombre', 'N/A')}, Edad: {top_result.get('Edad', 'N/A')}, Equipo: {top_result.get('Equipo', 'N/A')}, Pierna: {top_result.get('Pierna Buena', 'N/A')}, Posición: {top_result.get('Posición', 'N/A')}, Goles: {top_result.get('Goles', 'N/A')}, Asistencias: {top_result.get('Asistencias', 'N/A')}, Porterías a cero: {top_result.get('Porterías a Cero', 'N/A')}, Situación: {top_result.get('Situación', 'N/A')}, Proyección: {top_result.get('Proyección', 'N/A')}, Comentarios del Ojeador: {top_result.get('Comentarios del Ojeador', 'N/A')}"

    return search_result
```
